```python
#task1

class Calculator:
    def __init__(self):
        print("Let's Calculate")

    def calculate(self, first_value, operator, second_value):
        print('value1: ',first_value)
        print('Operator: ',operator)
        print('value2: ',second_value)

        if operator == "+":
            print('Result:',self.add(first_value,second_value)) 

        elif operator == "-":
            print('Result:',self.subtract(first_value,second_value))

        elif operator == "*":
            print('Result:',self.multiply(first_value,second_value))

        else:
            print('Result:',self.divide(first_value,second_value))
       

    def add(self, first_value,second_value):
        return first_value + second_value
      
    def subtract(self,first_value,second_value):
        return first_value - second_value

    def multiply(self,first_value,second_value):
        return first_value*second_value

    def divide(self,first_value,second_value):
        return first_value/second_value

        
num1 = int(input("first value : "))
operator = input("please enter your operator : ")
num2 = int(input("second value : "))
realcalcu = Calculator()
realcalcu.calculate(num1, operator,num2)
```

```python
#task2

class Customer:
    def __init__(self, name):
        self.name = name

    def greet(self,name = None):
        if name != None:
            print("Hello",name+"!")
        else:
            print("Hello!")
        
    
    def purchase(self, *args):
        items = args
        print(self.name+", you purchased",len(items),"item(s): ")
     
        for i in items:
            print(i)
 
customer_1 = Customer("Sam") 
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice") 
print("-----------------------------") 
customer_2 = Customer("David") 
customer_2.greet("David") 
customer_2.purchase("orange juice") 
 
```

```python
#task3

class Panda:

    def __init__(self,name,gender,age):
        self.name = name
        self.gender = gender
        self.age = age

    def sleep(self,sleep = None):

        if sleep == None:
            return '{} sleeps duration is unknown thus should have bamboo leaves'.format(self.name)

        elif sleep >=3 and sleep <=5:
            return '{} sleeps {} and should have Mixed Veggies'.format(self.name,sleep)

        elif sleep >= 6 and sleep <= 8:
            return '{} sleeps {} and should have Eggplant & Tofu'.format(self.name,sleep)
    
        elif sleep >= 9 and sleep <= 11:
            return '{} sleeps {} and should have Broccoli Chicken'.format(self.name,sleep)




panda1 = Panda("Kunfu","Male", 5)
panda2=Panda("Pan Pan","Female",3) 
panda3=Panda("Ming Ming's","Female",8) 
 
print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age)) 
 
print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age)) 
 
print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))      
print("===========================")      
print(panda2.sleep(10))      
print(panda1.sleep(4))      
print(panda3.sleep()) 
```

```python
#task13

class Batsman:
    def __init__(self,*args):
        if len(args) == 2:
            self.name = "New Batsman"
            self.scored = args[0]
            self.faced = args[1]


        elif len(args) == 3:
            self.name = args[0]
            self.scored = args[1]
            self.faced = args[2]



    def printCareerStatistics(self):
        print("NAME:", self.name)

        print(f'Runs Scored: {self.scored}, Balls Faced: {self.faced}')
        #print('Runs Scored:',self.scored,',Balls Faced:',self.faced)
        
        
    def battingStrikeRate(self):

        total = (self.scored/self.faced)*100

        return total

    def setName(self,name):
        self.name = name 

b1 = Batsman(6101, 7380) 
b1.printCareerStatistics() 
print("============================") 
b2 = Batsman("Liton Das", 678, 773) 
b2.printCareerStatistics() 
print("----------------------------") 
print(b2.battingStrikeRate()) 
print("============================") 
b1.setName("Shakib Al Hasan") 
b1.printCareerStatistics() 
print("----------------------------") 
print(b1.battingStrikeRate()) 
```

```python
#task16

class Author:
    def __init__(self,name = "Default",*args):
        self.name = name 


        self.book_name = []
        self.addBooks(*args)

    def addBooks(self,*args):

        for i in args:
            self.book_name.append(i)

    def changeName(self,name):
        self.name = name 


    def printDetails(self):
        print("Author Name:", self.name)
        print("-------- ")
        print("List of Books:")

        for i in self.book_name:
            print(i)


auth1 = Author('Humayun Ahmed') 
auth1.addBooks('Deyal', 'Megher Opor Bari') 
auth1.printDetails() 
print('===================') 
auth2 = Author() 
print(auth2.name) 
auth2.changeName('Mario Puzo') 
auth2.addBooks('The Godfather', 'Omerta', 'The Sicilian') 
print("===================") 
auth2.printDetails() 
print("===================") 
auth3 = Author('Paolo Coelho', 'The Alchemist', 'The Fifth Mountain') 
auth3.printDetails() 
```
