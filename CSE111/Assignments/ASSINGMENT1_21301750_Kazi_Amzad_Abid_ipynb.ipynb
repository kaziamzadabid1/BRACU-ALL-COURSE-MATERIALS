Two words are anagrams if they contain all of the same letters, but in a different order. For example, “evil” and “live” are anagrams because each contains one “e”, one “i”, one “l”, and one “v”. 
Write a program that reads two strings from the user and determines whether or not they are anagrams. Use a dictionary to solve the problem.

Sample Input: evil 
             live

Sample Output:  Those strings are anagrams

```python
user = input("please enter your first word : ")
user2 = input("please enter your second word : ")

a_dict = {}
a_dict2 = {}

for ch in user:
    if ch not in a_dict:
        a_dict[ch] = 1
    else:
        a_dict[ch] += 1

for chr in user2:
    if chr not in a_dict2:
        a_dict2[chr]=1
    else:
        a_dict2[ch] +=1

counter = 0

for i in a_dict:
    if a_dict[i] == a_dict2[i]:
        counter += 1 
#print(counter)
       

if counter == len(a_dict):
    print("Those strings are anagrams")
else:
    print("Those strings not are anagrams")
```
