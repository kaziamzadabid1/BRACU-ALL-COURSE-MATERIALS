```python
#task1

class Student:
    ID = 0
    def __init__(self,name,department, age,cgpa):
        self.name = name 
        self.department = department
        self.age = age 
        self.cgpa = cgpa 
        Student.ID += 1

    def get_details(self):
        print("ID",Student.ID)
        print("Name:",self.name)
        print("Departmetn:",self.department)
        print("Age",self.age)
        print("Cgpa",self.cgpa)
        

    @classmethod
    def from_String(cls,info):
        name, department, age,cgpa = info.split("-")  #ORStudent

        myobj = cls(name,department,age,cgpa)

        return myobj

s1 = Student("Samin", "CSE", 21, 3.91) 
s1.get_details() 
print("-----------------------") 
s2 = Student("Fahim", "ECE", 21, 3.85) 
s2.get_details() 
print("-----------------------") 
s3 = Student("Tahura", "EEE", 22, 3.01) 
s3.get_details()  
print("-----------------------") 
s4 = Student.from_String("Sumaiya-BBA-23-3.96") 
s4.get_details() 


print("\nA class variable can be accessed from anywhere by calling the class name with the variable. Besides, instance variables can’t be accessed from other classes if we have multiple classes. We usually use “self” to use the instance variable from a single class program.  \n")


print('An instance method is always created under a totally dependable class method; we use “self” as a parameter to track down all the info or the different memory location. On the other hand, Class instance is not required in class, the class method creates another class ,and we use “cls” as a parameter instead of “self.”' ) 
```

```python
#task2

class Assassin:
    counter = 0
    def __init__(self,name,rate):
        self.name = name 
        self.rate = rate
        Assassin.counter += 1 
        
    def printDetails(self):
        print("Name:", self.name)
        print("Success rate:", str(self.rate)+"%")
        print("Total number of Assassin:",Assassin.counter)

    @classmethod
    def failureRate(cls,name,info_rate):
        success_rate = 100-info_rate                        #20 
        obj = Assassin(name,success_rate)

        return obj

    @classmethod
    def failurePercentage(cls,name,info_percent):
        success_percent = 100-info_percent
        new_obj = cls(name,success_percent)
        return new_obj

john_wick = Assassin('John Wick', 100) 
john_wick.printDetails() 
print('================================') 
nagisa = Assassin.failureRate("Nagisa", 20) 
nagisa.printDetails() 
print('================================') 
akabane = Assassin.failurePercentage('Akabane', 10) 
akabane.printDetails()
```

```python
#task5

class Employee:
    def __init__(self,name,workingPeriod):
        self.name = name 
        self.workingPeriod = workingPeriod

    @classmethod
    def employeeByJoiningYear(cls,name,year): 

        obj = cls(name,2021-year)
        return obj


    @staticmethod
    def experienceCheck(workingPeriod,gender):
        result = ""

        if gender == "male":
            result += "He"
        else:
            result += "She"

        if workingPeriod <= 3:
             result += " is not experienced"
        else:
           result += " is experienced"

        return result


employee1 = Employee('Dororo', 3) 
employee2 = Employee.employeeByJoiningYear('Harry', 2016) 
print(employee1.workingPeriod) 
print(employee2.workingPeriod) 
print(employee1.name) 
print(employee2.name) 
print(Employee.experienceCheck(2, "male")) 
print(Employee.experienceCheck(3, "female"))
```

```python
class Cat:
    Number_of_cats = 0
    def __init__(self,color,action):
        self.color = color 
        self.action = action 
        Cat.Number_of_cats += 1

    @classmethod
    def no_parameter(cls,color = "white",action = "sitting",): 
        obj1 = cls(color,action)
        return obj1

    @classmethod
    def first_parameter(cls,color, action = "sitting"):
         obj2 = Cat(color,action)
         
         return obj2

    @classmethod
    def second_parameter(cls,action,color = "Grey"):
         obj3 = Cat(color,action)

         return obj3


    def printCat(self):
        print(self.color+" cat is "+self.action)


    def changeColor(self,colors):
        self.color = colors

print("Total number of cats:", Cat.Number_of_cats) 
c1 = Cat.no_parameter() 
c2 = Cat.first_parameter("Black") 
c3 = Cat("Brown", "jumping") 
c4 = Cat("Red", "purring") 
c5 = Cat.second_parameter("playing") 
print("=======================") 
c1.printCat() 
c2.printCat() 
c3.printCat() 
c4.printCat() 
c5.printCat() 
c1.changeColor("Blue") 
c3.changeColor("Purple") 
c1.printCat() 
c3.printCat() 
print("=======================") 
print("Total number of cats:", Cat.Number_of_cats)
```

```python
import math

class Cylinder:

    radius = 18
    height = 5

    def __init__(self,height,radius):
        print("Default radius=",Cylinder.height ,"and height=",Cylinder.radius)

        print("Updated: radius",height,"and updated:height=",radius)


        Cylinder.radius = radius                                 #formula of cylinder  = pi*r**h
        Cylinder.height = height

    @classmethod
    def swap(cls,height,radius):
        myobj1 = Cylinder(radius,height)

        return myobj1


    @classmethod
    def changeFormat(cls,info):
        radius,height = info.split("-")
        radius = float(radius)
        height = float(height)

        myobj = Cylinder(radius,height)
        return myobj
        

    @staticmethod
    def area(height,radius):
        area = 2*math.pi*radius**2 + 2*math.pi*radius*height
        print("Area:", area)
        
    @staticmethod
    def volume(height,radius):
         
         volume = math.pi*radius**2*height
         print("Volume:",volume)

c1 = Cylinder(0,0) 
Cylinder.area(c1.radius,c1.height) 
Cylinder.volume(c1.radius,c1.height) 
print("===============================") 
c2 = Cylinder.swap(8,3) 
c2.area(c2.radius,c2.height) 
c2.volume(c2.radius,c2.height) 
print("===============================") 
c3 = Cylinder.changeFormat("7-13") 
c3.area(c3.radius,c3.height) 
c3.volume(c3.radius,c3.height) 
print("===============================") 
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height) 
print("===============================") 
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
import math

class SultansDine:

    num_of_branch = 0
    total = 0
    mylist = []
    
    def __init__(self,location):
        self.location = location 
        self.branch_sell = 0                     #store100000
        SultansDine.num_of_branch += 1
        SultansDine.mylist.append(self)
    
    def sellQuantity(self,quantity):
        if quantity <10:
            self.branch_sell += quantity*300

        elif quantity <20:
             self.branch_sell += quantity*350

        else:
            self.branch_sell += quantity*400

        
        SultansDine.total += self.branch_sell

    
    def branchInformation(self):
        print("Branch Name:",self.location)
        print("Branch sell:",self.branch_sell,"Taka" )

    
    @classmethod
    def details(cls):
        print("Total Number of branch(s):",cls.num_of_branch) 
        
        print("Total Sell:",cls.total,"Taka")

        for i in cls.mylist:
            print("Branch Name: {}, Branch Sell: {} Taka ".format(i.location,i.total))

            ori = (i.branch_sell/cls.total)*100
            print("Branch consists of total sell's:",round(ori,2),"%")


SultansDine.details() 
print('########################') 
dhanmodi = SultansDine('Dhanmondi') 
dhanmodi.sellQuantity(25) 
dhanmodi.branchInformation() 
print('-----------------------------------------') 
SultansDine.details() 

print('========================') 

baily_road = SultansDine('Baily Road') 
baily_road.sellQuantity(15)
baily_road.branchInformation() 
print('-----------------------------------------') 
SultansDine.details() 
 
print('========================') 
 
gulshan = SultansDine('Gulshan') 
gulshan.sellQuantity(9) 
gulshan.branchInformation() 
print('-----------------------------------------') 
SultansDine.details()
```
