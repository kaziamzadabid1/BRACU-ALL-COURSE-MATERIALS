```python
#TASK4

class Color:
    def __init__(self,color = None):
        self.clr = color 

    def __add__(self,other_color):
         
         if (self.clr == "red" and other_color.clr == "yellow") or (self.clr == "yellow" and other_color.clr == "red"):
             return Color("Orange")

         elif self.clr == "red" and other_color.clr == "blue" or self.clr == "blue" and other_color.clr == "red" :
             return Color("Violet")
             
         elif self.clr == "yellow" and other_color.clr == "blue" or self.clr == "blue" and other_color.clr == "yellow" :
            return Color("Green")

         elif self.clr == other_color.clr and (self.clr  == "red" or self.clr == "blue" or self.clr == "yellow"):
            return Color(self.clr) 

         else:
             return Color("Invalid")  

C1 = Color(input("First Color: ").lower()) 
C2 = Color(input("Second Color: ").lower()) 
C3 = C1 + C2 
print("Color formed:", C3.clr) 
```

```python
#TASK5

import math

class Circle:
    def  __init__(self,radius):
        self.__radius = radius


    def setRadius(self,radius):
        self.__radius = radius


    def getRadius(self):
        return self.__radius
        
    def area(self):
        total_area = (math.pi*self.__radius**2)
        return total_area 
        

    def __add__(self,new_radius):
        c3 = self.__radius + new_radius.__radius

        return Circle(c3)

c1 = Circle(4)  
print("First circle radius:" , c1.getRadius()) 
print("First circle area:" ,c1.area()) 

c2 = Circle(5) 
print("Second circle radius:" ,c2.getRadius()) 
print("Second circle area:" ,c2.area()) 

c3 = c1 + c2  
print("Third circle radius:" ,c3.getRadius()) 
print("Third circle area:" ,c3.area())
```

```python
#TASK6

class Triangle:
    def __init__(self,base,height):
        self.__base = base 
        self.__height = height

    def setBase(self,base):
        self.__base = base


    def getBase(self):
        return self.__base
      

    def setHeight(self,height):
        self.__height = height


    def getHeight(self):
        return self.__height

    def area(self):
        total_areaoftri = ((1/2)*(self.__base*self.__height))

        return total_areaoftri

    def __sub__(self,other):
        totl_base = self.__base - other.__base 
        total_height = self.__height - other.__height 
        

        return Triangle( totl_base,total_height)

t1 = Triangle(10, 5) 
print("First Triangle Base:" , t1.getBase()) 
print("First Triangle Height:" , t1.getHeight()) 
print("First Triangle area:" ,t1.area()) 
 
t2 = Triangle(5, 3) 
print("Second Triangle Base:" , t2.getBase()) 
print("Second Triangle Height:" , t2.getHeight()) 
print("Second Triangle area:" ,t2.area()) 
 
t3 = t1 - t2  
print("Third Triangle Base:" , t3.getBase()) 
print("Third Triangle Height:" , t3.getHeight()) 
print("Third Triangle area:" ,t3.area()) 
```

```python
#TASK7

class Dolls:
    def __init__(self,name,price, merged = False):
        self.name = name
        self.price = price
        self.merged = merged                                     # True

    def __gt__(self,others):
        if self.price > others.price:
            return True 
        else:
            return False

    def detail(self):

        if self.merged:  
            return "Dolls: "+self.name +"\n"+"Total Price: "+ str(self.price)+" taka"
        else:
             return "Doll: "+self.name+"\n"+"Total Price: "+ str(self.price)+" taka"


    def __add__(self,others):
        addition_name = self.name + " "+others.name
        addition_price = self.price + others.price

        return Dolls(addition_name,addition_price,True)

obj_1 = Dolls("Tweety", 2500) 
print(obj_1.detail()) 
if obj_1 > obj_1:     
    print("Congratulations! You get the Tweety as a gift!") 
else:     
    print("Thank you!") 
 
print("=========================") 
obj_2 = Dolls("Daffy Duck", 1800)  
print(obj_2.detail()) 
if obj_2 > obj_1:     
    print("Congratulations! You get the Tweety as a gift!") 
else:     
    print("Thank you!") 
 
print("=========================") 
obj_3 = Dolls("Bugs Bunny", 3000) 
print(obj_3.detail()) 
if obj_3 > obj_1:     
    print("Congratulations! You get the Tweety as a gift!") 
else:     
    print("Thank you!") 
 
print("=========================") 
obj_4 = Dolls("Porky Pig", 1500) 
print(obj_4.detail()) 
if obj_4 > obj_1:     
    print("Congratulations! You get the Tweety as a gift!") 
else:     
    print("Thank you!") 
 
print("=========================") 
obj_5 = obj_2 + obj_3 
print(obj_5.detail()) 
if obj_5 > obj_1:     
    print("Congratulations! You get the Tweety as a gift!") 
else:     
    print("Thank you!") 
```

```python
#TASK8 
class Coordinates:
    def __init__(self,num1,num2):
        self.num1 = num1
        self.num2 = num2


    def __sub__(self,numbers):
        subresult = self.num1 - numbers.num1    # X
        subresult2 = self.num2 - numbers.num2   # Y

        return Coordinates(subresult,subresult2)
        

    def __mul__(self,numbers):
        multiresult = self.num1 * numbers.num1
        multisult2 = self.num2 * numbers.num2

        return Coordinates(multiresult,multisult2)

    def __eq__(self,numbers):
        if self.num1 == numbers.num1 and self.num2 == numbers.num2:          #X and Y coordinates 
            return "The calculated coordinates are the same."
        else:
            return "The calculated coordinates are NOT the same."

    def detail(self):
        #return "("+str(self.num1)+","+str(self.num2)+")"
        return (self.num1, self.num2)

p1 = Coordinates(int(input()),int(input())) 
p2 = Coordinates(int(input()),int(input())) 
 
p4 = p1 - p2  
print(p4.detail()) 
 
p5 = p1 * p2 
print(p5.detail()) 
 
point_check = (p4 == p5) 
print(point_check)
```
