 Write a function which takes only your full name as input and finds the total number of vowels in that input. Output every vowel and at the end output the total number of vowels found. If your name does not contain any vowel then your function should print “No vowels in the name

 Sample Input (Steve Jobs) 

 OUTPUT : Vowels: e, e, o. Total number of vowels: 3 

```python
def namefunc(name):

    string = "aeiou"

    vowel_counter = ''

    mituation = name.lower()
    

    for chr in mituation:
        if chr in string:
            vowel_counter += chr
            
    for i in range (len(vowel_counter)):
        if i == (len(vowel_counter)-1):
            print(vowel_counter[i],end = '.')
        else:
            print(vowel_counter[i],end = ',')
    print(f' Total number of vowels:{len(vowel_counter)}')





user_name = input("please enter your name : ")
namefunc(user_name)
```

```python
def calender(days):


    year = days//365

    remain = days%365

    month = remain//30

    day = remain%30 

    print('{} year, {} months and {} days'.format(year,month,day))

user = int(input("please enter your days : "))

calender(user)
```
